```python
import torch
import torch.nn.functional as F
import numpy as np
import networkx as nx
import torchvision

import matplotlib.pyplot as plt
```

```python
# Load self.base
# model = .load_state_dic(torch.load("first_trained_base.pth"))

# Load Cifar10 dataset
dataset = torchvision.datasets.CIFAR10(root = "./data/cifar10/", train = True)

xs = torch.load("xs.tensor")
ys = torch.load("ys.tensor")
idxs = torch.load("idxs.tensor")
features = torch.load("features.tensor") # normalized output features
print("xs: {}\nys: {}\nidxs: {}\nfeatures: {}".format(xs.shape, ys.shape, idxs.shape, features.shape))
```

```python
def visualize(indices, x, y):
    """indices on torchvision.datasets.CIFAR10
    """
    #plt.close('all')
    fig, axes = plt.subplots(x,y, figsize = (10,10))
    for i in range(y):
        for j in range(x):
            axes[i][j].set_axis_off()
            axes[i][j].imshow(dataset[indices[(i*y)+j]][0])
```

```python
y_mask = (ys==torch.unique(ys)[0]) # y : frog (6: frog, 7:horse)

random_index = np.random.choice(idxs[y_mask], 25, replace = False)

visualize(random_index, 5, 5)
```

```python
y_mask = y_mask.cpu()
feature = features[y_mask]
idx = (idxs.cpu()).numpy()[y_mask]

_sim_matrix = torch.relu(F.cosine_similarity(feature.unsqueeze(1), feature.unsqueeze(0), dim=-1))

#print(feature.unsqueeze(1).shape) => (250, 1, 256)
#print(feature.unsqueeze(0).shape) => (1, 250, 256)
sim_matrix = _sim_matrix.type(torch.float32)
sim_matrix[sim_matrix == 0] = 1e-5

sim_matrix = sim_matrix.cpu().numpy()

g = nx.from_numpy_matrix(sim_matrix)
info = nx.eigenvector_centrality(g, max_iter=6000, weight = 'weight')
centrality = np.array([info[i] for i in range(len(info))])
```

```python
print(feature.unsqueeze(1).shape)
print(feature.unsqueeze(0).shape)
```

```python
sorted_order = np.argsort(centrality)[::-1]
sorted_index = idx[sorted_order]

visualize(sorted_index[:25], 5, 5)
```

```python
centroid_idx = np.argmax(centrality)
sim_matrix[centroid_idx].shape
```

```python
norm_centrality = (centrality - centrality.min()) / (centrality.max() - centrality.min())
norm_sim_matrix = (sim_matrix - sim_matrix.min()) / (sim_matrix.max() - sim_matrix.min())
print(centrality.max())
print(centrality.min())
print(norm_centrality.max())
print(norm_centrality.min())
print(norm_sim_matrix.max())
print(norm_sim_matrix.min())
```

```python

alpha = 0.5
print("alpha is {}".format(alpha))

#score = ((alpha) * norm_centrality) + ((1-alpha) * (-norm_sim_matrix[centroid_idx]))
score = ((alpha) * centrality) + ((1-alpha) * (-sim_matrix[centroid_idx]))

sorted_order = np.argsort(score)[::-1]
sorted_index = idx[sorted_order]
visualize(sorted_index[:25], 5, 5)
```
